```python
import turtle
wn=turtle.Screen()
wn.title('pong game by sachin dangi')
wn.bgcolor('purple')
wn.setup(width=800,height=600)
wn.tracer(0)

# for score of both players
score_a=0
score_b=0

# paddle A 
paddle_a=turtle.Turtle()
paddle_a.speed(0)
paddle_a.shape('square')
paddle_a.color('red')
paddle_a.shapesize(stretch_wid=5,stretch_len=1)
paddle_a.penup()
paddle_a.goto(-350,0)

# paddle B
paddle_b=turtle.Turtle()
paddle_b.speed(0)
paddle_b.shape('square')
paddle_b.color('green')
paddle_b.shapesize(stretch_wid=5,stretch_len=1)
paddle_b.penup()
paddle_b.goto(350,0)

# Ball
ball=turtle.Turtle()
ball.speed(0)
ball.shape('circle')
ball.color('yellow')
ball.penup()
ball.goto(0,0)
# for movement of ball
ball.dx=1
ball.dy=-1

# pen for writing the score of each player 
pen =turtle.Turtle()
pen.speed(0)
pen.color('green')
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write('player A: 0  player B: 0',align='center',font=('courier',24,'normal'))

# function for upward movement of paddle A 
def paddle_a_up():
    y=paddle_a.ycor()
    y += 20
    paddle_a.sety(y)

# function for downward movement of paddle A 
def paddle_a_down():
    y=paddle_a.ycor()
    y -= 20
    paddle_a.sety(y)
    
# function for upward movement of paddle B
def paddle_b_up():
    y=paddle_b.ycor()
    y += 20
    paddle_b.sety(y)
    
# function for downward movement of paddle B
def paddle_b_down():
    y=paddle_b.ycor()
    y -= 20
    paddle_b.sety(y)
    
    
# for controlling movement of paddle A  
wn.listen()
# for upward movement 
wn.onkeypress(paddle_a_up,'w')
# for downward movement
wn.onkeypress(paddle_a_down,'s')

# for controlling movements of paddle B
wn.onkeypress(paddle_b_up,'Up')
# for downward movement of paddle B
wn.onkeypress(paddle_b_down,'Down')




# main loop of game
while True:
    wn.update()
    ball.setx(ball.xcor()+ball.dx)
    ball.sety(ball.ycor()+ball.dy)
    
    # Border checking 
    if ball.ycor()>290:
        ball.sety(290)
        ball.dy*=-1
        
    if ball.ycor()<-290:
        ball.sety(-290)
        ball.dy*=-1
        
    if ball.xcor()>390:
        ball.goto(0,0)
        ball.dx*=-1
        #for score 
        score_a+=1
        pen.clear()
        pen.write('player A: {}  player B: {}'.format(score_a,score_b,align='center',font=('courier',24,'normal'))
        
    if ball.xcor()<-390:
        ball.goto(0,0)
        ball.dx*=-1
        #for score 
        score_b+=1
        pen.clear()          
        pen.write('player A: {}  player B: {}'.format(score_a,score_b,align='center',font=('courier',24,'normal'))
        
    # paddle and ball collisions
    if(ball.xcor()>340 and ball.xcor()<350) and (ball.ycor()<paddle_b.ycor()+40 and ball.ycor()>paddle_b.ycor()-40):
        ball.setx(340)
        ball.dx*=-1
        
        
    if(ball.xcor()<-340 and ball.xcor()>-350) and (ball.ycor()<paddle_a.ycor()+40 and ball.ycor()>paddle_a.ycor()-40):
        ball.setx(-340)
        ball.dx*=-1
    
```
